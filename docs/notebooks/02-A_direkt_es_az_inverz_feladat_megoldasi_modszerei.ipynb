## Példa a dolgozatban használt adathalmazok elemeire

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

import numpy as np
```

```python
# AFHQ
# az adathalmaz elérhető az alábbi linken: https://www.kaggle.com/datasets/andrewmvd/animal-faces
data_dir = r'./datas/afhq/train'
image_size = (512, 512)

dataset = keras.preprocessing.image_dataset_from_directory(
    data_dir, image_size=image_size, batch_size=32
)

class_names = ["cat", "dog", "wild"]

plt.figure(figsize=(8, 4), dpi=100)
for images, labels in dataset.take(1):
    for i in range(8):
        ax = plt.subplot(2, 4, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
plt.show()
```

```python
#Cifar-10

(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()

class_names = ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]

plt.figure(figsize=(8, 4), dpi=100)

for i in range(10):
    index = np.where(y_train == i)[0][0]
    ax = plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[index], interpolation="none")
    plt.title(class_names[int(y_train[index])])
    plt.axis("off")
plt.show()
```
