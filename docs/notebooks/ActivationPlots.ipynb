# Aktivációs függvények plot

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
def plot_activation(x, y, text_x, text_y, formula_string):
    fig = plt.figure(figsize=(7, 3), dpi=300)
    axes = fig.add_axes([0, 0, 1, 1])

    # 5pts 2 spaces 1 pt 2 spaces
    axes.grid(True, color='0.6', dashes=(5, 2, 1, 2))

    axes.spines['left'].set_position('zero')
    axes.spines['right'].set_visible(False)
    axes.spines['bottom'].set_position('zero')
    axes.spines['top'].set_visible(False)
    axes.plot((1), (0), ls="", marker=">", ms=8,
              color="k", transform=axes.get_yaxis_transform(), clip_on=False)
    axes.plot((0), (1), ls="", marker="^", ms=8,
              color="k", transform=axes.get_xaxis_transform(), clip_on=False)

    axes.yaxis.set_major_locator(ticker.MultipleLocator(0.5))

    axes.plot(x, y, linewidth=2)

    axes.text(text_x, text_y,
              formula_string, size='xx-large',
              bbox=dict(boxstyle='round', facecolor='wheat'))
    
    plt.show()
```

## Hyperbolic Tangent

```python
x = np.linspace(-2.5, 2.5, 1000)
tanh = ( 2 / (1 + np.exp(-2*x) ) ) -1

plot_activation(x, tanh, -2.5, 0.8, r'$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$')
```

## Sigmoid

```python
x = np.linspace(-6, 6, 1000)
sigmoid = 1 / (1 + np.exp(-x) )

plot_activation(x, sigmoid, -6, 0.8, r'$\sigma(x) = \frac{1}{1 + e^{-x}}$')
```

## ReLU

```python
x = np.linspace(-2, 2, 1000)
relu = np.maximum(0, x)

plot_activation(x, relu, -2, 1.7, r'$f(x) = \max(0, x)$')
```

# Leaky ReLU

```python
x = np.linspace(-2, 2, 1000)
leaky_relu = np.maximum(0.1 * x, x)

plot_activation(x, leaky_relu, -2, 1.7, r'$f(x) = \max(0.1x, x)$')
```
