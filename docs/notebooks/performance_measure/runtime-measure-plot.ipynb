```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
laptop = np.load("../datas/runtimemeasures/sizes_avg_runtime_laptop.npy")
print(laptop)

kaggle = np.load("../datas/runtimemeasures/sizes_avg_runtime_kaggle.npy")
print(kaggle)

colab = np.load("../datas/runtimemeasures/sizes_avg_runtime_colab.npy")
print(colab)
```

```python
fig = plt.figure(figsize=(8, 4), dpi=300)
ax = fig.add_axes([0, 0, 1, 1])
#plt.xscale("log", base=2)

x = [0, 1, 2, 3]

ax.plot(x, laptop, color='b', alpha=.75, lw=2, ls='-.', marker='v', markersize=10,
        markerfacecolor='b', markeredgecolor='b', markeredgewidth=2, label="Dell Insprion-5558")

ax.plot(x, colab, color='m', alpha=.75, lw=2, ls='--', marker='h', markersize=10,
        markerfacecolor='m', markeredgecolor='m', markeredgewidth=2, label="Colab környezet")

ax.plot(x, kaggle, color='g', alpha=.75, lw=2, ls='--', marker='o', markersize=10,
        markerfacecolor='g', markeredgecolor='g', markeredgewidth=2, label="Kaggle környezet")

ax.legend(loc=0, handlelength=5)
ax.grid(True, color='0.6', dashes=(5, 2, 1, 2))

ax.set_title('Átlagos epoch futásidő a felbontásnövelő rétegek függvényében')
ax.set_ylabel('Futásidő (s)')
ax.set_xlabel('Felbontásnövelő rétegek száma\n(kimeneti kép felbontása)')

labels = ["0\n(4 x 4)", "1\n(8 x 8)", "2\n(16 x 16)", "3\n(32 x 32)"]
plt.xticks(x, labels, rotation='horizontal')

plt.show()
```
