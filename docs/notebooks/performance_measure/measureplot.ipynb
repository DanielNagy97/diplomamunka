```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
laptop = np.load("../datas/runtimemeasures/sizes_avg_runtime_laptop.npy")
print(laptop)

kaggle = np.load("../datas/runtimemeasures/sizes_avg_runtime_kaggle.npy")
print(kaggle)

colab = np.load("../datas/runtimemeasures/sizes_avg_runtime_colab.npy")
print(colab)
```

```python
fig = plt.figure(figsize=(8, 4), dpi=300)
ax = fig.add_axes([0, 0, 1, 1])
plt.xscale("log", base=2)

x = [4, 8, 16, 32]

ax.plot(x, laptop, color='b', alpha=.75, lw=2, ls='-.', marker='v', markersize=10,
        markerfacecolor='b', markeredgecolor='b', markeredgewidth=2, label="Dell Insprion-5558")

ax.plot(x, kaggle, color='g', alpha=.75, lw=2, ls='--', marker='o', markersize=10,
        markerfacecolor='g', markeredgecolor='g', markeredgewidth=2, label="Kaggle krnyezet")

ax.plot(x, colab, color='y', alpha=.75, lw=2, ls='--', marker='h', markersize=10,
        markerfacecolor='y', markeredgecolor='y', markeredgewidth=2, label="Colab krnyezet")

ax.legend(loc=0, handlelength=5)
ax.grid(True, color='0.6', dashes=(5, 2, 1, 2))

ax.set_title('tlagos epoch futsid a felbonts fggvnyben')
ax.set_ylabel('Futsid (s)')
ax.set_xlabel('Felbonts')

labels = ["(4 x 4)", "(8 x 8)", "(16 x 16)", "(32 x 32)"]
plt.xticks(x, labels, rotation='horizontal')

plt.show()
```
